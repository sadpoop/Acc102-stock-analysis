```python
import matplotlib.pyplot as plt
import pandas as pd


df_all = pd.read_csv("stock_data.csv")
df_all["Date"] = pd.to_datetime(df_all["Date"], format="%Y-%m-%d")


print("Available columns:", df_all.columns.tolist())


fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

stocks = df_all['company_name'].unique()
for i, stock in enumerate(stocks):
    data = df_all[df_all['company_name'] == stock]
   
    axes[i].plot(data['Date'], data['Close'], label=stock, linewidth=1.5)
    axes[i].set_title(f'{stock} Close Price')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Price (USD)')
    axes[i].legend()
    axes[i].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('close_price_subplots.png', dpi=300)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, stock in enumerate(stocks):
    data = df_all[df_all['company_name'] == stock]
    axes[i].bar(data.index, data['Volume'], label=stock, alpha=0.6, color='orange')
    axes[i].set_title(f'{stock} Trading Volume')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Volume')
    axes[i].legend()
    axes[i].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('volume_subplots.png', dpi=300)
plt.show()
```

```python
for stock in stocks:
    data = df_all[df_all['company_name'] == stock].copy()
    
    # 用你CSV里真实存在的 Close 列，而不是 Adj Close
    data['MA10'] = data['Close'].rolling(10).mean()
    data['MA20'] = data['Close'].rolling(20).mean()
    data['MA50'] = data['Close'].rolling(50).mean()

    plt.figure(figsize=(12, 6))
    # 用 Date 列当x轴，而不是 data.index，避免索引问题
    plt.plot(data['Date'], data['Close'], label='Close', alpha=0.7)
    plt.plot(data['Date'], data['MA10'], label='10-Day MA', linewidth=1.2)
    plt.plot(data['Date'], data['MA20'], label='20-Day MA', linewidth=1.2)
    plt.plot(data['Date'], data['MA50'], label='50-Day MA', linewidth=1.2)
    
    plt.title(f'{stock} Price with Moving Averages')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.savefig(f'{stock}_moving_averages.png', dpi=300)
    plt.show()
```

```python
df_all['Daily Return'] = df_all.groupby('company_name')['Adj Close'].pct_change()

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, stock in enumerate(stocks):
    data = df_all[df_all['company_name'] == stock]
    axes[i].hist(data['Daily Return'].dropna(), bins=30, alpha=0.7, color='purple')
    axes[i].set_title(f'{stock} Daily Return Distribution')
    axes[i].set_xlabel('Daily Return')
    axes[i].set_ylabel('Frequency')
    axes[i].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('daily_return_histograms.png', dpi=300)
plt.show()
```

```python
import seaborn as sns
import numpy as np


df_all['Daily_Return'] = df_all.groupby('company_name')['Adj Close'].pct_change()


company_list = [df_all[df_all['company_name'] == name].copy() for name in df_all['company_name'].unique()]
tech_name = df_all['company_name'].unique().tolist()
tech_list = tech_name
```

```python
plt.figure(figsize=(15,10))
for i, com in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    com['Daily_Return'].plot(lw=1.2)
    plt.title(f"Daily Return: {tech_name[i-1]}")
    plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("4_Daily_Return_Line.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i, com in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    com['Daily_Return'].hist(bins=50, alpha=0.7)
    plt.title(f"Return Distribution: {tech_name[i-1]}")
plt.tight_layout()
plt.savefig("5_Return_Hist.png", dpi=300)
plt.show()
```

```python
close_df = df_all.pivot_table(
    values='Close',        
    index='Date',          
    columns='company_name'
)


ret_df = close_df.pct_change()


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(ret_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Stock Return Correlation Heatmap")
plt.savefig("6_Correlation_Heatmap.png", dpi=300)
plt.show()
```
